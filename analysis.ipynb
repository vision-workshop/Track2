```python
import numpy as np
from PIL import Image
import glob
import json
import matplotlib.pyplot as plt 
import os
```

```python
# load image path as well as annotations based on which we want to generate more synthetic defect

res_file = "./work_dirs/Lenses_sliced/20230223_015735.log.json"  # path to the run log file
```

```python
def res_log_analysis(res_file):
    res_log = []
    for line in open(res_file, 'r'):
        res_log.append(json.loads(line))
    train_loss = []
    val_mAP = []
    for i in range(len(res_log)):
        try:
            if res_log[i]['mode']=='train':
                train_loss.append(res_log[i]['loss'])
            if res_log[i]['mode']=='val':
                val_mAP.append(res_log[i]['segm_mAP_50'])
        except:
            pass
    return train_loss, val_mAP
```

```python
train_loss, val_mAP = res_log_analysis(res_file) # retrive the train_loss log and val_mAP50 log 
```

```python
plt.plot(train_loss,  label = 'train_loss')
plt.legend()
# plt.ylim([0, 0.20])
```

```python
plt.plot(val_mAP, label = 'val_mAP_slice')
plt.legend()
# plt.ylim([0, 0.20])
```
